```python
# 2. Which country has won the most gold medals in summer games?
df1.index[df1.Gold==max(df1.Gold[0:146])].tolist()[0]
```

```python
# 3. Which country had the biggest difference between their summer and winter gold medal counts?
m=0
i=-1
for index in range(0,len(df1.index)-2,1):
    temp=abs(df1["Gold"][index]-df1["Gold.1"][index])
    if temp>m:
        m=temp
        i=index
df1.index[i]  
```

```python
# Which country has the biggest difference between their summer gold medal counts and winter gold medal counts relative to their total gold medal count? 
# Only include countries that have won at least 1 gold in both summer and winter.
m=0
i=-1
for index in range(0,len(df1.index)-2,1):
    if df1["Gold"][index]>0 and df1["Gold.1"][index]>0:
        temp=abs(df1["Gold"][index]-df1["Gold.1"][index])/(df1["Gold"][index]+df1["Gold.1"][index])
        if temp>m:
            m=temp
            i=index
df1.index[i]
```

```python
# Write a function that creates a Series called "Points" which is a weighted value where each gold medal (Gold.2) counts for 3 points,
#silver medals (Silver.2) for 2 points, and bronze medals (Bronze.2) for 1 point. 
#The function should return only the column (a Series object) which you created, 
#with the country names as indices.(This function should return a Series named Points of length 146)
def getSeries(df1):
    country_names=df1.index
    points=df1["Gold.2"]*3+df1["Silver.2"]*2+df1["Bronz.2"]*1
    s=pd.Series(points,name="Points")
    s.index=country_names
    return s
```
