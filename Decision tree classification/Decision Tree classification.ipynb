### Content
This dataset includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms in the Agaricus and Lepiota Family Mushroom drawn from The Audubon Society Field Guide to North American Mushrooms (1981). Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; no rule like "leaflets three, let it be'' for Poisonous Oak and Ivy.

### Encoding

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
```

```python
for col in df1.columns:
    print("----"+col+"-----")
    print(df1[col].unique())
    df1[col]=le.fit_transform(df1[col])
    print(df1[col].unique())
```

```python
df1.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
Y1=df1['class']
cols1=df1.columns.tolist()[1:23]
X1=df1[cols1]
```

```python
x_train1,x_test1,y_train1,y_test1=train_test_split(X1,Y1,test_size=0.2,random_state=42)
```

```python
acc_list1=[]
for depth in range(1,20):
    model1=DecisionTreeClassifier(max_depth=depth)
    model1.fit(x_train1,y_train1)
    ypred1=model1.predict(x_test1)
    acc_list1.append(accuracy_score(y_test1,ypred1))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,20),acc_list1,color="blue",linestyle="dashed",marker="o",markerfacecolor="red",markersize=10)
plt.title("accuracy vs. depth value")
plt.xlabel("depth value")
plt.ylabel("accuracy")
plt.show()
```

```python
depth1=acc_list1.index(max(acc_list1))
```

```python
model2=DecisionTreeClassifier(max_depth=depth1+1)
model2.fit(x_train1,y_train1)
ypred2_test=model2.predict(x_test1)
accuracy_score(y_test1,ypred2_test)
```

```python
ypred2_train=model2.predict(x_train1)
accuracy_score(y_train1,ypred2_train)
```
