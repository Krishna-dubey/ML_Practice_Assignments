```python
from matplotlib import pyplot as plt
from matplotlib import style
```

Basic plot

```python
plt.plot([2,3,5],[6,3,7])
plt.show()
```

Graph title and axis label

```python
x=[2,3,5]
y=[6,3,7]
plt.plot(x,y)
plt.title("Testing matplotlib")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

Styling graph using ggplot

```python
style.use('ggplot')
x = [1,2,3]
y = [4,5,6]
x2 = [7,8,9]
y2 = [10,11,12]
plt.plot(x,y,'blue',label='line one', linewidth=5)
plt.plot(x2,y2,'red',label='line two',linewidth=5)
plt.title('Test Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.legend(['first line','second line'])
plt.grid(True,color='y')
plt.show()
```

# Types of plot:
1. Bar Graph
2. Histogram
3. Scatter Plot
4. Area Plot
5. Pie Chart

### 1. Bargraph
- used for comparing data

```python
plt.bar([0.5,2.0,3.5,5,6.5],[50,40,70,80,20],
label="label1",width=1,color='g')
plt.bar([1.25,2.75,4.25,5.75,7.25],[80,20,20,50,60],
label="label2", color='r',width=0.5)
plt.legend()
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('Bargraph')
plt.show()
```

### 2. Histogram
- Gives the distribution of data

```python
ages=[2,3,4,13,20,23,25,26,28,28,30,45,55]
plt.hist(ages,bins=[0,10,20,30,40,50],color='brown',edgecolor='black')
plt.xlabel('age')
plt.ylabel('count')
plt.title('Age Distribution Histogram')
plt.show()
```

### 3. Scatter plot

```python
x = [1,2,3]
y = [4,5,6]
 
x1=[7,8,9]
y1=[10,11,12]
 
plt.scatter(x,y, label='cluster 1',color='green')
plt.scatter(x1,y1,label='cluster 2',color='purple')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

### 4. Area plot

```python
days = [1,2,3,4,5]
  
sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[],color='m', label='Sleeping', linewidth=5)
plt.plot([],[],color='c', label='Eating', linewidth=5)
plt.plot([],[],color='r', label='Working', linewidth=5)
plt.plot([],[],color='k', label='Playing', linewidth=5)

plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack Plot')
plt.legend()
plt.show()
```

```python
slices=[1,1,1,1]
plt.pie(slices,
       labels=['first','second','third','fourth'],
       colors=['green','yellow','red','blue'],
       shadow=True,
        autopct='%0.1f%%',
       explode=(0,0.1,0,0.25))
```
